```python
import matplotlib.pyplot as plot
import pandas as pd
data = pd.read_csv('planets.csv')
```

```python
fig = plot.figure()
plot.plot(data['pl_radj'], data['pl_bmassj'],'.k')
fig.suptitle('Mass-Radius Relationship of Exoplanets')
plot.xlabel('Radius (RJ)')
plot.ylabel('Mass (MJ)')
```

```python
plot.savefig('mass-radius')
```
