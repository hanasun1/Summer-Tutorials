```python
import matplotlib.pyplot as plot
import pandas as pd
data = pd.read_csv('planets (1).csv')
```

```python
fig = plot.figure()
plot.plot(data['pl_orbper'], data['pl_orbsmax'],'k.')
fig.suptitle('Exoplanet: Period vs Semi Major Axis')
plot.xlabel('Orbital Period (days)')
plot.ylabel('Orbital Semi-Major Axis (AU)')
```
